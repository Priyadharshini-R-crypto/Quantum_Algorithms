```python
from qiskit import QuantumCircuit
from qiskit.circuit import Gate
from math import pi
qc = QuantumCircuit(2)
c = 0
t = 1
```

```python
# a controlled-Z
qc.cz(c,t)
qc.draw()
```

```python
qc = QuantumCircuit(2)
# also a controlled-Z
qc.h(t)
qc.cx(c,t)
qc.h(t)
qc.draw()
```

```python
qc = QuantumCircuit(2)
# a controlled-Y
qc.sdg(t)
qc.cx(c,t)
qc.s(t)
qc.draw()
```

```python
qc = QuantumCircuit(2)
# a controlled-H
qc.ry(pi/4,t)
qc.cx(c,t)
qc.ry(-pi/4,t)
qc.draw()
```

```python
a = 0
b = 1

qc = QuantumCircuit(2)
# swaps states of qubits a and b
qc.swap(a,b)
qc.draw()
```

```python
qc = QuantumCircuit(2)
# swap a 1 from a to b
qc.cx(a,b) # copies 1 from a to b
qc.cx(b,a) # uses the 1 on b to rotate the state of a to 0
qc.draw()
```

```python
# swap a q from b to a
qc.cx(b,a) # copies 1 from b to a
qc.cx(a,b) # uses the 1 on a to rotate the state of b to 0
qc.draw()
```

```python
qc = QuantumCircuit(2)
theta = pi # theta can be anything (pi chosen arbitrarily)
qc.ry(theta/2,t)
qc.cx(c,t)
qc.ry(-theta/2,t)
qc.cx(c,t)
qc.draw()
```

```python
A = Gate('A', 1, [])
B = Gate('B', 1, [])
C = Gate('C', 1, [])
alpha = 1 # arbitrarily define alpha to allow drawing of circuit
```

```python
qc = QuantumCircuit(2)
qc.append(C, [t])
qc.cz(c,t)
qc.append(B, [t])
qc.cz(c,t)
qc.append(A, [t])
qc.p(alpha,c)
qc.draw()
```

```python
from qiskit import QuantumCircuit
from qiskit.quantum_info import Statevector
from qiskit.visualization import plot_bloch_vector
import numpy as np
import matplotlib.pyplot as plt

# Bloch vector function
def get_bloch_vector(qc):
    state = Statevector.from_instruction(qc)
    a = state.data[0]
    b = state.data[1]
    x = 2 * (a.conjugate() * b).real
    y = 2 * (a.conjugate() * b).imag
    z = abs(a)**2 - abs(b)**2
    return [x, y, z]

# Step 1: Initial |0⟩
qc0 = QuantumCircuit(1)
bloch_vec0 = get_bloch_vector(qc0)
display(plot_bloch_vector(bloch_vec0, title="Step 1: |0⟩"))

# Step 2: RY(π/4)
qc1 = QuantumCircuit(1)
qc1.ry(np.pi / 4, 0)
bloch_vec1 = get_bloch_vector(qc1)
display(plot_bloch_vector(bloch_vec1, title="Step 2: After RY(π/4)"))

# Step 3: RY(π/4) → X
qc2 = QuantumCircuit(1)
qc2.ry(np.pi / 4, 0)
qc2.x(0)               # Simulate controlled X
bloch_vec2 = get_bloch_vector(qc2)
display(plot_bloch_vector(bloch_vec2, title="Step 3: Simulating CH (RY → X → RY†)"))

# Step 3: RY(π/4) → X → RY(−π/4)
qc3 = QuantumCircuit(1)
qc3.ry(np.pi / 4, 0)
qc3.x(0)              # Simulate controlled X
qc3.ry(-np.pi / 4, 0)
bloch_vec3 = get_bloch_vector(qc3)
display(plot_bloch_vector(bloch_vec3, title="Step 3: Simulating CH (RY → X → RY†)"))


qc4 = QuantumCircuit(2)
qc4.ry(pi/4, 1)       # Prepare target (qubit 1)
qc4.cx(0, 1)          # Controlled-X: control=0, target=1
qc4.ry(-pi/4, 1)      # Undo the basis change
qc4.draw('mpl')
bloch_vec4 = get_bloch_vector(qc4)
display(plot_bloch_vector(bloch_vec4, title="Step 3: Simulating CH (RY → X → RY†)"))
```
